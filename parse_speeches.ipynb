```python
import json

with open(r"info_folder\author_maps.json", "r", encoding="utf-8") as f:
    author_map = json.load(f)
```

```python
total_remaining = 0

for row in author_map:
    author_name = row["author_key"]
    print(author_name)

    data = data_processing_utils.query_fed_prints_by_author(author_name)
    links, remaining = data_processing_utils.retrieve_remaining_ids(author_name, data)

    total_remaining += remaining

    print(links)

    author_short_name = data_processing_utils.get_author_short_name(author_name)

    for inst, link_by_inst in links.items():
        if inst == "pdf":
            data_processing_utils.pdfs_to_json(links["pdf"], f"text_data/{author_short_name}.json")
        else:
            data_processing_utils.html_speeches_to_json(link_by_inst, inst, f"text_data/{author_short_name}.json")

print(total_remaining)
```
